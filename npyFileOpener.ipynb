```python
import numpy as np

filePath = 'pathToFile'
data = np.load(filePath)
```

```python
import os

newFolderPathAll = 'folder to extract to'
```

```python
from PIL import Image

for index in range(data.shape[0]):
  imageData = data[index]
  image = Image.fromarray(imageData.astype(np.uint8))
  imgPath = os.path.join(newFolderPathAll, f'image_{index}.png')
  image.save(imgPath)
```
